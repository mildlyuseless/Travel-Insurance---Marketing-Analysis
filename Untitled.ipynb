# Travel Insurance - Marketing Analysis (Practice Case Study)

### INTRODUCTION

This notebook contains the case study of travel insurance from Travel Assured! Travel Assured is a travel insurance
company. Due to the COVID pandemic, they have had to cut their marketing budget by over 50%. It is more important than
ever that they advertise in the right places and to the right people.

Travel Assured has plenty of data on their current customers as well as people who got quotes but never bought insurance.
They want to know if there are differences in the travel habits between customers and non-customers - they believe they
are more likely to travel often (buying tickets from frequent flyer miles) and travel abroad. Our business task is to
uncover any insights from these data that might help the marketing strategy.

### ASK

From our business task, we can ask the following questions:

 - **What are the differences in the travel habits between customers and non-customers?**

 - **How can these travel habits be turned into marketing opportunities?**

#### PREPARE

Loading in the required libraries and importing the data. It should be noted that throughout this notebook, the terms 'customer' and 'insured' are used interchangeably.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('travel_insurance.csv')
df.head()
```

```python
df.info()
```

#### OBSERVATION
Looking at the data provided, it appears fairly well labelled and organized. Some tidying can be done to make the data easier to work with. It also appears that many of the variables are categorical.

The data will be encoded with indicator values to increase readability. Some variable names will also be changed to increase readability, while keeping the default naming scheme. 

```python
df_dummy = pd.get_dummies(df, drop_first=True)

# Rename columns for readability; attempted to keep the same naming convention used
df_dummy.columns = ['Age', 'AnnualIncome(k$)', 'FamilyMembers',
                    'HealthIssues', 'Insured', 'PrivateSector',
                    'Graduate', 'FrequentFlyer', 'TraveledAbroad']

# Converting AnnualIncome to be in k$
df_dummy['AnnualIncome(k$)'] = df_dummy['AnnualIncome(k$)'] / 1000
```

```python
# Data after column rename
df_dummy.describe().round(3)
```

```python
cols = ['HealthIssues', 'PrivateSector', 'Graduate', 'FrequentFlyer', 'TraveledAbroad']

for i in cols:
    a = df_dummy[df_dummy[i] == 1]
    b = a.Insured.value_counts(normalize=True).round(3)
    print('{} customer rate: {}%'.format(i, b[1] * 100))
```

#### OBSERVATION
Now that the data has been cleaned, further exploration can begin. Based on the data, 35.7% of travelers are customers (have Travel Insurance). To gain a better understanding of the differences between customers and non-customers (insured vs. uninsured), the data set will be split by 'Insured' status.

Some interesting values indicate that the initial assumption that those that have traveled abroad and those that are frequent flyers are more likely to be customers.

```python
sns.heatmap(df_dummy.corr(), annot=True)
plt.title('Correlation Heatmap - All')

plt.tight_layout()
plt.savefig('Correlation Heatmap - All.png')
```

#### OBSERVATION
The data shows that there are differences in the following categories between customers and non-customers:

 - Annual Income
 - Private Sector
 - Frequent Flyer
 - Travel Abroad
 
There is little to no significant correlation between any other categories.

#### NEXT STEPS:
As the correlation matrix above confirms, Annual Income is the key area of interest. Although age did not initially appear to be significant, the data needed to be checked and some interesting results appeared. Customer and non-customer data frames will be coded for ease of use.

```python
cust = df_dummy[df_dummy['Insured'] == 1].drop(columns='Insured')
non_cust = df_dummy[df_dummy['Insured'] == 0].drop(columns='Insured')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(11,4), sharey=True)
fig.suptitle('Ages of  Non-customers vs. Customers')

sns.countplot(x='Age', data=non_cust, ax=ax[0])
ax[0].set(xlabel='Age - Non-Customer')

sns.countplot(x='Age', data=cust, ax=ax[1])
ax[1].set(xlabel='Age - Customer')

plt.tight_layout()
plt.savefig('cust_vs_non_age.png')
```

```python
twenty_8 = df_dummy[df_dummy['Age'] == 28].drop(columns='Age')
twenty_8 = (twenty_8.Insured.value_counts(normalize=True) * 100).round(2)

print('{}% of 28 year old travelers are not customers'.format(twenty_8[0]))
```

```python
non_cust_28 = non_cust[non_cust['Age'] == 28].drop(columns='Age')
non_cust_28.describe().round(3)
```

#### OBSERVATION
28 year old traveler's appear to be the most frequent age group in this data set. They also happen to be the most frequent age group that tends not to purchase insurance. Taking a look at the averages of 28 year old traveler's that are not customers, it appears that the average annual income is less than the data set as a whole.

28 year old travelers earn, on average, approximately $155,000 less than the entire data set. One other area to be noted is that they are highly educated with an almost 96% graduate rate, low travel abroad rate of 4%, and only 14% are frequent flyers.

#### NEXT STEPS:
Given the large discrepancy in average annual incomes between the 28 year old population and the data as a whole, further exploration of the distribution of annual income is needed.

```python
plt.hist('AnnualIncome(k$)', data=cust, label='Customer', alpha=0.4)
plt.hist('AnnualIncome(k$)', data=non_cust, label='Non-customer', alpha=0.4)

plt.title('Annual Income by Customer')
plt.xlabel('Annual Income (Thousands of $)')
plt.ylabel('Count')
plt.legend()

plt.tight_layout()
plt.savefig('income_by_cust.png')
```

```python
over_130k = df_dummy[df_dummy['AnnualIncome(k$)'] > 1300]
cust_over = (over_130k.Insured.value_counts(normalize=True) * 100).round(2)

under_130k = df_dummy[df_dummy['AnnualIncome(k$)'] < 1300]
cust_under = (under_130k.Insured.value_counts(normalize=True) * 100).round(2)


print('{}% of travelers that earn over $1.3million annually are customers'.format(cust_over[1]))
print('{}% of travelers that earn under $1.3million annually are not customers'.format(cust_under[0]))
```

#### OBSERVATION
The distribution of annual income by customer status is very noticeable. There is a clear distinction in annual income between the two groups. Travelers with an annual income less than $1.3million were high unlikely to purchase travel insurance.

The scatterplot below further illustrates that there is a clear gap between annual income and the purchase of travel insurance. This plot also shows that there is no correlation between age and customer status.

```python
# Annual income by insured status scatterplot
ax = sns.scatterplot(x='Age', y='AnnualIncome(k$)', hue='Insured', data=df_dummy, alpha=0.4)
plt.title('Age vs Income by Insured Status')
plt.ylabel('Annual Income (Thousands of $)')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.savefig('age_income_scatterplot.png')
```

#### NEXT STEPS:
Visualization of the other identified areas of interest: traveled abroad, frequent flyers, and private sector.

```python
sns.catplot(x='TraveledAbroad', y='AnnualIncome(k$)', data=df_dummy, hue='Insured', alpha=0.4)
plt.xticks([0, 1], ['No', 'Yes'])
plt.suptitle('Income distribution by Travel and Insurance')

plt.tight_layout()
plt.savefig('abroad_insured_income.png')
```

```python
sns.catplot(x='FrequentFlyer', y='AnnualIncome(k$)', data=df_dummy, hue='Insured', alpha=0.4)
plt.xticks([0, 1], ['No', 'Yes'])
plt.suptitle('Income distribution by Frequent flyers and Insurance')

plt.tight_layout()
plt.savefig('freqfly_insured_income.png')
```

```python
sns.catplot(x='PrivateSector', y='AnnualIncome(k$)', data=df_dummy, hue='Insured', alpha=0.3)
plt.xticks([0, 1], ['No', 'Yes'])
plt.suptitle('Income distribution by Private Sector and Insurance')

plt.tight_layout()
plt.savefig('privsec_insured_income.png')
```

```python
sns.countplot(x='PrivateSector', data=df_dummy, hue='Insured', alpha=0.4)
plt.xticks([0, 1], ['No', 'Yes'])
plt.suptitle('Income distribution by Private Sector and Insurance')

plt.tight_layout()
plt.savefig('cust_private_sec_count.png')
```

```python
gov_sec = df_dummy[(df_dummy['PrivateSector'] == 0)]
gov_sec = 1 - (gov_sec.Insured.value_counts(normalize=True)).round(3)
print('{}% of Government sector are not customers'.format((gov_sec[1] * 100)))

priv_sec = df_dummy[(df_dummy['PrivateSector'] == 1)]
priv_sec = 1 - (priv_sec.Insured.value_counts(normalize=True)).round(3)
print('{}% of Private sector are not customers'.format(priv_sec[1] * 100))
```

#### OBSERVATION
While frequent flyers and those that have traveled abroad show a larger gap betwen incomes, the distinction between income and the private sector is not as clear. However, when looking at the customer rates between the two sectors below, the Government sector has a much lower customer rate than the private sector.

## CONCLUSION
Based on the data provided, it appears that the initial assessment made by Travel Assured is correct. Customers tend to be frequent flyers and are more likely to have traveled abroad. In addition, customers have, on average, a higher income greater than $1,300,000 annually. The data shows that wealthier travelers, regardless of age, number of family members, or health issues tend to be more risk averse which seems to make sense.

Non-customers have a lower annual income, compared to customers, at less than $1,300,000 annually; 76.92% of travelers are not customers. The age distribution of non-customers shows a large area of opportunity to target 28 year old travelers; 79.25% of 28 year old travelers are not customers. The final area of interest lies within the Government sector population of travelers. The success rate of insured Government sector travelers is only 24.6% compared to the public sector/self-employed 40.2%.
